### Cleaning Data

```python
df_all[df_all.isna().any(axis=1)]
```

```python
num_cols = ['PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3', 'TEMP', 'PRES', 'DEWP', 'RAIN', 'WSPM']
df_all[num_cols] = df_all[num_cols].fillna(df_all[num_cols].median())
```

```python
df_all['wd'].value_counts()
```

```python
df_all['wd'].fillna(value="NE", inplace=True)
```

```python
df_all.isna().sum()
```

```python
thresholds = {
    'PM2.5': df_all['PM2.5'].quantile(0.95),
    'PM10': df_all['PM10'].quantile(0.95),
    'SO2': df_all['SO2'].quantile(0.95),
    'NO2': df_all['NO2'].quantile(0.95),
    'CO': df_all['CO'].quantile(0.95),
    'O3': df_all['O3'].quantile(0.95),
}

for col, max_val in thresholds.items():
    df_all[col].replace(df_all[col].max(), max_val, inplace=True)
```

```python
df_all.describe()
```

**Insight:**
- Ditemukan 38.600 baris data dengan missing values, terutama pada variabel PM2.5, PM10, NO2, CO, dan O3.
- Missing values pada variabel numerik diisi dengan nilai median masing-masing kolom.
- Missing values pada variabel kategori 'wd' (arah angin) diisi dengan nilai modus ('NE') karena merupakan kategori yang paling sering muncul.
- Setelah imputasi, tidak ada lagi nilai yang hilang dalam dataset (df_all.isna().sum() menunjukkan 0 pada semua kolom).
- Untuk mengatasi outlier, nilai maksimum untuk PM2.5, PM10, SO2, NO2, CO, dan O3 disesuaikan dengan persentil 95% untuk menghindari distorsi analisis.

## Visualization & Explanatory Analysis

### Pertanyaan 1:

```python
import matplotlib.pyplot as plt
import seaborn as sns

city_stations = ["Dongsi", "Guanyuan", "Tiantan", "Nongzhanguan", "Aotizhongxin", "Wanshouxigong", "Wanliu"]
suburb_stations = ["Changping", "Dingling", "Huairou", "Shunyi", "Gucheng"]

ordered_stations = city_stations + suburb_stations

avg_pollution_per_station = df_all.groupby("station")[["PM2.5", "PM10", "NO2"]].mean()

avg_pollution_per_station = avg_pollution_per_station.loc[ordered_stations]

plt.figure(figsize=(14, 6))
ax = avg_pollution_per_station.plot(kind="bar", colormap="coolwarm", alpha=0.85, edgecolor="black")

plt.axvline(x=len(city_stations) - 0.5, color="black", linestyle="--", linewidth=2)

plt.text(len(city_stations) / 2 - 1, avg_pollution_per_station.max().max() * 1.1, "Pusat Kota",
         ha="center", fontsize=12, fontweight="bold", color="blue")
plt.text(len(city_stations) + len(suburb_stations) / 2 - 1, avg_pollution_per_station.max().max() * 1.1, "Pinggiran",
         ha="center", fontsize=12, fontweight="bold", color="green")

plt.title("Rata-rata Konsentrasi Polutan per Stasiun (2013-2017)", pad=20, fontsize=14)

plt.legend(title="Jenis Polutan", bbox_to_anchor=(1.05, 1), loc="upper left")

plt.ylabel("Konsentrasi Polutan (µg/m³)")
plt.xlabel("Stasiun Pengukuran")
plt.xticks(rotation=45, ha="right")

plt.grid(axis="y", linestyle="--", alpha=0.7)

plt.ylim(0, avg_pollution_per_station.max().max() * 1.2)

plt.tight_layout()
plt.show()
```

### Pertanyaan 2:

```python
import seaborn as sns
import matplotlib.pyplot as plt

mean_pollutants = df_all.groupby("time_of_day")[["PM2.5", "PM10", "NO2"]].mean().reset_index()

plt.figure(figsize=(10, 6))
colors = ["#ff9999", "#66b3ff", "#99ff99"]

mean_pollutants.set_index("time_of_day").plot(kind="bar", color=colors, alpha=0.8)
plt.title("Rata-rata Konsentrasi Polutan (Siang vs Malam)")
plt.xlabel("Waktu")
plt.ylabel("Konsentrasi (µg/m³)")
plt.xticks(rotation=0)
plt.legend(["PM2.5", "PM10", "NO2"])
plt.grid(axis="y", linestyle="--", alpha=0.7)

plt.show()
```

### Pertanyaan 3:

```python
import seaborn as sns
import matplotlib.pyplot as plt

corr_values = df_all[["PM2.5", "PM10", "NO2", "RAIN", "WSPM"]].corr().iloc[:3, 3:]

plt.figure(figsize=(8, 6))
corr_values.plot(kind="bar", colormap="coolwarm", alpha=0.85, edgecolor="black")

plt.axhline(y=0, color="gray", linestyle="--", alpha=0.7)
plt.title("Korelasi Curah Hujan & Kecepatan Angin terhadap Polutan")
plt.ylabel("Korelasi")
plt.xticks(rotation=0)
plt.legend(title="Variabel Cuaca")
plt.grid(axis="y", linestyle="--", alpha=0.7)

plt.show()
```
